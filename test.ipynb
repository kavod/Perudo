```python
import PerudoGame as pg
import PerudoExcept as pe
import NNK2 as NNK
import math
import numpy as np
import random
import matplotlib.pyplot as plt
MAX_PLAYERS=6
MAX_DICES = 5
```

```python
score = [0]*6
NB_PLAYERS = random.randint(3,6)
game = pg.PerudoGame(NB_PLAYERS)
sit = None
while sum(score)<100:
    try:
        prev_sit = sit
        play = random.choice(game.possiblePlay())
        sit = game.dices[game.turn]
        sit = np.append(sit,sum(game.nb_dices))
        sit = np.append(sit,[play['nb'],play['val'],int(play['no']),int(play['pile'])])
        print(sit)
        game.play(nb=play['nb'],val=play['val'],no=play['no'],pile=play['pile'])
    except pe.PerudoWinner as e:
        #print(e.args[1])
        score[e.args[1]] = score[e.args[1]] + 1
        NB_PLAYERS = random.randint(3,6)
        game = pg.PerudoGame(NB_PLAYERS)
plt.plot(score)
```
